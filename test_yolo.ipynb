```python
import matplotlib.pyplot as plt
import glob
from yolo_model import detect_objects, draw_boxes
# load test images
test_images = [plt.imread(path) for path in glob.glob('./test_images/*.jpg')]

# plot
rows = int(len(test_images)/2)
fig, ax = plt.subplots(ncols=2,nrows=rows)
fig.set_size_inches((20,20))

for i in range(rows):
    for j in range(2):
        img = test_images[i*2 + j]
        bbox, labels = detect_objects(img)
        ax[i,j].imshow(draw_boxes(img, bbox, labels))
```
